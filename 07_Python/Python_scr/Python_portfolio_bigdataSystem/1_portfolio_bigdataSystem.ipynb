## 1. yes24의 베스트셀러 정보를 제공하는 사이트에서 베스트셀러 정보를 수집해서 파일에 저장하세요.
베스트셀러 정보 수집 주소 : http://www.yes24.com/24/category/bestseller

### 화면 출력하고 txt파일로 저장

```python
import requests
from urllib.request import urlopen
from bs4 import BeautifulSoup
url = 'http://www.yes24.com/24/category/bestseller'
# best_seller = requests.get(url)
# soup = BeautifulSoup(best_seller.content, 'html.parser')
bestseller = urlopen(url)
soup = BeautifulSoup(bestseller, "html.parser")
titles = soup.select('div > ol > li > p:nth-child(3) > a')
authors = soup.select('ol > li > p.aupu')
prices = soup.select('ol > li > p.price > strong')

with open ('data/1_yes24.txt','w', encoding='utf-8') as f:
    print('순위,책이름,저자및출판사,가격')
    f.write('순위,책이름,저자및출판사,가격\n')
    for idx in range(len(titles)):
        print('{},{},{},{}'.format(idx+1, titles[idx].text, authors[idx].text, prices[idx].text))
        f.write('{},{},{},{}\n'.format(idx+1, titles[idx].text, authors[idx].text, prices[idx].text))
```

### 딕셔너리를 데이터프레임으로 받아서 csv로 저장

```python
import pandas as pd
import requests
from urllib.request import urlopen
from bs4 import BeautifulSoup
url = 'http://www.yes24.com/24/category/bestseller'
bestseller = urlopen(url)
titles = soup.select('div > ol > li > p:nth-child(3) > a')
authors = soup.select('ol > li > p.aupu')
prices = soup.select('ol > li > p.price > strong')

bestsel_list = [] # 빈 리스트
for idx in range(len(titles)):
    dic = {}
    dic['순위'] = idx+1
    dic['책이름'] = titles[idx].text
    dic['저자및출판사'] = authors[idx].text
    dic['가격'] = prices[idx].text
    bestsel_list.append(dic)
df = pd.DataFrame(bestsel_list)
df.to_csv('data/1_yes24_1.csv', encoding='cp949', index=False)
df
```
