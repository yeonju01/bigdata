# 연습문제
## 2. 다음에제시된웹페이지는많이본순으로선정한Daum뉴스의랭킹페 이지입니다(http://media.daum.net/ranking/popular). 이 페이지에서 각 뉴스의 제목과 이 뉴스를 올린 신문사명을 스크래핑(50개)하여 newstitle, newscomname 형식의 데이터 프레임을 만들고 news.csv라 는 csv파일로 저장하는 프로그램을 구현하여 제출합니다.

```python
import requests
from bs4 import BeautifulSoup
url = 'http://media.daum.net/ranking/popular'
news_ranking = requests.get(url)
soup = BeautifulSoup(news_ranking.content, 'html.parser')
title_list = soup.select('div.cont_thumb strong.tit_thumb a.link_txt')
dic_list1 = [] # 빈 리스트
for idx in range(len(title_list)):
    dic = {} # 빈 딕셔너리
    dic['newstitle'] = title_list[idx].text
    dic_list1.append(dic)
dic_list1
```

```python
import requests
from bs4 import BeautifulSoup
url = 'http://media.daum.net/ranking/popular'
news_ranking = requests.get(url)
soup = BeautifulSoup(news_ranking.content, 'html.parser')
title_list = soup.select('div.cont_thumb strong.tit_thumb a.link_txt')
comname = soup.select('div.cont_thumb strong.tit_thumb span')
dic_list2 = [] # 빈 리스트
for idx in range(len(comname)):
    dic = {} # 빈 딕셔너리
    dic['newscomname'] = comname[idx].text
    dic_list2.append(dic)
dic_list2
```

```python
import pandas as pd
d = dic_list1
df1 = pd.DataFrame(data=d)
df1
```

```python
import pandas as pd
d = dic_list2
df2 = pd.DataFrame(data=d)
df2
```

```python
df = pd.concat([df1, df2], axis=1)
df.to_csv('data/2_news.csv', encoding='utf-8', index=False)
df
```
